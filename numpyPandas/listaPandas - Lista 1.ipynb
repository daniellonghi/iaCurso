```python
#1 Do arquivo carros, imprima as primeiras e as últimas cinco linhas. 
#Faça uma versão que abre o arquivo a partir do csv e outra que abre a partir da url.

import pandas as pd

url = "https://raw.githubusercontent.com/advinstai/python-datascience/master/Exercicios/carros.csv"

#READ FROM URL
#relampagoMarquinhos = pd.read_csv("carros.csv")
relampagoMarquinhos = pd.read_csv(url)
print("Primeiros 5 registros")
display(relampagoMarquinhos.head(5))
print("-----------------\n\nUltimos 5 registros")
display(relampagoMarquinhos.tail(5))
```

```python
relampagoMarquinhos = open("carros.csv", "r")
relampagoMarquinhos = relampagoMarquinhos.readlines()
headerList = []
valuesList = []

for i in range(len(relampagoMarquinhos)):
    if i == 0:
        headerList = relampagoMarquinhos[i].rstrip().split(",")
    else:
        valuesList.append(relampagoMarquinhos[i].rstrip().split(","))

#SHOW PRIMEIROS 5
print("Os primeiros 5")
for i in range(5):
    if i == 0:
        print(headerList)
        print(valuesList[i])
    else:
        print(valuesList[i])

#SHOW ULTIMOS 5
print("\n-----------------\n\nOs ultimos 5")
print(headerList)
for i in range(len(valuesList)-1, len(valuesList)-6,-1):
    print(valuesList[i])
```

```python
#2 Encontre o nome da empresa de carros mais cara
relampagoMarquinhos = pd.read_csv(url)
display(relampagoMarquinhos)
priceCarro = relampagoMarquinhos["price"].max()
retorno = relampagoMarquinhos.loc[relampagoMarquinhos["price"] >= priceCarro]
print(retorno["company"])
```

```python
#3 mostre a média de preços
print(relampagoMarquinhos["price"].mean())
```

```python
#4 mostre o valor mais alto e mais baixo de horsepower
print(relampagoMarquinhos["horsepower"].max())
print(relampagoMarquinhos["horsepower"].min())
```

```python
#5 mostre a informação de todos os carros da marcar toyota
print(relampagoMarquinhos.loc[relampagoMarquinhos["company"] == "toyota"])
```

```python
#6 faça uma função que filtra os carros por cilindro e salva em outro arquivo CSV
def filtraPorCilindrada(cilindrada):
    return relampagoMarquinhos.loc[relampagoMarquinhos["num-of-cylinders"] == cilindrada]
    

retornoFiltro = filtraPorCilindrada("four")
retornoFiltro.to_csv(r'csvByCylinders.csv')

print(pd.read_csv("csvByCylinders.csv"))
```
