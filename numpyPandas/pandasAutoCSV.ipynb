```python
import pandas as pd

url = "https://raw.githubusercontent.com/advinstai/python-datascience/master/csv/Auto.csv"

df = pd.read_csv(url, na_values="?")
```

```python
df.isnull().sum()
```

```python
#Remove os registros sem informacoes
df = df.dropna()

#Preenche com um valor desejado os valores que sao nulos ou vazios
df = df.fillna()

#Busca por um valor especifico
display(df[df["horsepower"] == "?"])

display(df)
```

```python
valorMedia = df["horsepower"].mean()

df = df.fillna(valorMedia)

display(df)

df.isnull().sum()
```
