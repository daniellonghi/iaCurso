Modele uma classe que realiza as seguintes operações em um arquivo texto

```python
import statistics
from fractions import Fraction as F

class textToList():
    __stopWords = []
    __textoList = ""
    
    #Cadastre StopWords (A classe deve possuir um atributo com uma lista de StopWords)
    def setStopWords(self,stopWord):
        self.__stopWords.append(stopWord)
    
    def getStopWords(self):
        return self.__stopWords
    
    def setTexto(self,texto):
        self.__textoList = texto
    
    def getTexto(self):
        return self.__textoList
    
    #Represente um texto como uma lista de Strings
    #Retorne individualmente cada palavra do texto
    def eachWord(self):
        listaRetorno = self.__textoList.split(" ")
        return listaRetorno
    
    #Conte a quantidade de ocorrências de cada palavra do texto
    def occurrencyWord(self):
        retornoDic = {}
        eachWord = self.eachWord()
        for i in range(len(eachWord)):
            if eachWord[i].lower() in retornoDic:
                retornoDic[eachWord[i].lower()]+=1
            else:
                retornoDic[eachWord[i].lower()] = 1
            
        return retornoDic
    
    #Retorne as 10 palavras mais frequentes
    def topTen(self, tipo):
        eachWord = self.occurrencyWord()
        tempWords = sorted(eachWord.items() , reverse=True, key=lambda x: x[1])[:10]
        tmpReturn = []
        if tipo == "key":
            tmpReturn = ([(key) for (key, value) in tempWords])
        else:
            tmpReturn = ([(value) for (key, value) in tempWords])
        return tmpReturn
    
    #Retorne a média e desvio padrão da quantidade de ocorrências
    def meanStdD(self):
        itensRetorno = self.topTen("value")
        totalItens = 0
        for i in range(len(itensRetorno)):
            totalItens+= itensRetorno[i]
            
        strItem = statistics.stdev(itensRetorno)
        return "Mean: " + str(totalItens/len(itensRetorno)) + " - " +  "Std: " + str(strItem)
    
    #Retorne um novo arquivo eliminando todas as StopWords do texto
    def returnWithoutStop(self):
        returnText = self.__textoList
        for i in range(len(self.__stopWords)):
            returnText = returnText.replace(str(self.__stopWords[i]), "")
        return returnText
    
    #Inclua um método que retorne a distância entre duas palavras
    def returnLength(self, primeira, segunda):
        primeira = self.__textoList.find(primeira)
        segunda = self.__textoList.find(segunda)
        
        return segunda-primeira
    
a = textToList()
a.setTexto("This is thisd dhsahd a ad dsa a text a text to wthat to as 123 512 asd ghteh 234 examples.")
print(a.getTexto())

print(a.eachWord())
print(" ---- ")
for i in range(len(a.eachWord())):
    print (a.eachWord()[i])
print(" ---- aqui")
for values in a.occurrencyWord():
    print (values)
print(" ---- ")
print(a.occurrencyWord())
print(" ---- ")
print(a.topTen("key"))
print(" ---- ")
print(a.meanStdD())
print(" ---- ")
a.setStopWords("a")
a.setStopWords("t")
print(a.getStopWords())
print(" ---- ")
print(a.returnWithoutStop())
print(" ---- ")
print(a.returnLength("this","ad"))
```
