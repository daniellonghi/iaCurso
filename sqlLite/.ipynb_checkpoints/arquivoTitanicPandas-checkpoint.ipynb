```python
#1 - Leia os dados train.csv usando pandas
#2 - Remova a coluna com nomes de pessoas
#3 - A coluna Cabin contém uma letra e um número. Crie uma nova coluna chamada 
#    deck usando apenas a letra de cada registro presente na coluna Cabin.
#4 - Substitua os valores ausentes pela média de valores da coluna
#5 - Os valores da coluna "embarked" sao os seguintes: C = Cherbourg, Q = Queenstown, S = Southampton
#6 - Crie um CSV com essas uma coluna nome_cidade_de_embarque: C = Cherbourg, Q = Queenstown, 
#    S = Southampton e mais uma coluna chamada população com valores hipotéticos, faça um merge entre o CSV train.csv e esse novo CSV com os nomes das cidades. Implemente um filtro de passageiros por população da cidade de embarque

import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/advinstai/python-datascience/master/Exercicios/titanic/train.csv"

dataSet = pd.read_csv(url)
```

```python
print(dataSet.head())
print(dataSet.describe())
```

```python
display(dataSet)
dataSet = dataSet.drop(columns="Name")
display(dataSet)
```

```python
#dataSet = dataSet.dropna()
display(dataSet)
```

```python
dataSet["Deck"] = ""
display(dataSet)
i = 0
for index, row in dataSet.iterrows():
    if not (pd.isnull(row["Cabin"])):
        itemSplitted = row["Cabin"].split(" ")
        itensAdd = ""
        for i in range(len(itemSplitted)):
            itensAdd += itemSplitted[i][0:1] + " "
        dataSet.loc[index, "Deck"] = itensAdd

display(dataSet)
```

```python
for each in list(dataSet.describe().columns):
    dataSet.fillna({each:dataSet[each].mean()}, inplace=True)

display(dataSet)
```

```python
#Os valores da coluna "embarked" sao os seguintes: C = Cherbourg, Q = Queenstown, S = Southampton

dataSetCidade = pd.DataFrame(columns=['nome_cidade_de_embarque'])

dataSetCidade["Embarked"] = dataSet["Embarked"]
dataSetCidade["nome_cidade_de_embarque"] = dataSet["Embarked"]

dataSetCidade["nome_cidade_de_embarque"] = dataSetCidade["nome_cidade_de_embarque"].str.replace("C","Cherbourg", case = False)
dataSetCidade["nome_cidade_de_embarque"] = dataSetCidade["nome_cidade_de_embarque"].str.replace("Q","Queenstown", case = False)
dataSetCidade["nome_cidade_de_embarque"] = dataSetCidade["nome_cidade_de_embarque"].str.replace("S","Southampton", case = False)

display(dataSetCidade)

dataSetCidade.to_csv("CidadesEmbarque.csv", index = False)
```

```python
#RELENDO ARQUIVOS
display(dataSet)

dataSetCidade = pd.read_csv("CidadesEmbarque.csv")
display(dataSetCidade)
```

```python
dataSetMerged = dataSet.merge(dataSetCidade, left_on = "Embarked", right_on = "Embarked")

display(dataSetMerged)
```
